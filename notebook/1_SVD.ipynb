```python
import requests

from surprise import SVD
from surprise import Dataset, Reader
from surprise.model_selection import train_test_split
import pandas as pd
```

## Aquisição dos Dados

```python
path = 'https://raw.githubusercontent.com/' \
       + 'morsoletodev/RecommendationSystems/main/data/dados_treinamento.parquet'

df = pd.read_parquet(path)
```

```python
df['Peso'].replace({3:2, 5:3}, inplace=True)
```

```python
df.head()
```

## Preparativos

```python
# Parser de arquivo/df
reader = Reader(
    rating_scale=(1, 3)  # Precisamos determinar o intervalo da coluna Peso
)
```

```python
# Converte o pd.DataFrame para surprise.dataset.DatasetAutoFolds
data = Dataset.load_from_df(df[['ID: Usuário', 'ID: Jogo', 'Peso']], reader)
```

```python
# Separar dados para treinamento e teste
trainset, testset = train_test_split(data, test_size=0.25, random_state=42)
```

## Aplicação do Modelo

```python
# Instancia o modelo
svd = SVD(random_state=42)
```

```python
# Treinamento do Modelo
svd.fit(trainset)
```

```python
# Teste de valores
predictions_svd = svd.test(testset)

# É possível fazer predições individuais:
# pred_indiv_svd = svd.predict(uid=1, iid=103, r_ui=None, verbose=True)
```
